 ## Réponse Q3
$ f_{2} : \mathbb{R}^2 \rightarrow \mathbb{R}$, $ (x_1,x_2) \mapsto 100(x_2-x_1^2)^2 + (1-x_1)^2$

Justifier que l’algorithme puisse ne pas converger pour $f_{2}$ avec certains points initiaux.

Matrice hessienne de $f_{2} : \begin{pmatrix} 1212x_1^{2}-400x_2-4 & -400x_1\\ -400x_1 & 200 \\ \end{pmatrix}$

Les points initiaux pour lesquels l'agorithme ne converge pas, sont ceux qui ont une matrice hessienne dont le déterminant est nul (dans ce cas, la matrice est non-inversible et il n'y aura pas convergence avec Newton).
Si on résout l'équation du déterminant nul, on trouve : $y = \frac{103}{100}x^{2}-\frac{1}{100}$.
Ainsi pour tous les points qui vérifient cette équation (voir figure ci-dessous) l'algorithme ne peut pas converger.

Le point (0, -0.01) est a un facteur 2 près le même point que x023 (entre f2 des tests et f2 de la question il y a un facteur 2 devant le 2ème terme). Or on l'a vu au dessus, le test de Newton sur ce point sort avec le flag = 3 car on atteint le nombre d'iter max justement parce qu'il n'y a pas convergence.

```julia
using Plots

x = range(-1, 1, length=100)
f(x) = (103/100)*x^2 - 1/100
plot(x, f, title="Det(Hf2) = 0", label ="", linewidth=3)
ylims!(-0.05, 0.05)
xlims!(-0.5, 0.5)
xlabel!("x")
ylabel!("y")
scatter!([0], [-0.01], color = "green", label = "[0,-0.01]", markersize = 3)
```
